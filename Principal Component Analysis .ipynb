# Principal Component Analysis

PCA = unsupervized statistical technique used to examine the interrelations among a set of variables in order to identify the underlying structure of those variables (also know as general factor analysis)

where LR determines a line of best fit to a data set, factor analysis determines several orthogonal (at right angles, lines are parpendicular to each other in n-dimensional space)lines of best fit to the data set

N-dimensional space is the variable sample space - as many dimension as are variables

Components\
-are a linear transformation that chooses a variable system for the data set such that the greatest variance of the data set comes to lie on the fist axis\
-the second greatest variance on the second axis, and so on\
-this process allow us to reduce the number of variables used in an analysis\
-components are uncorrelated, since in the sample space they are orthgonal to each other

if we use this technique on a data set with a large number of variables, we can compress the amount of explained variations to just a few components

the most challenging part of PCA is the interpreting the components

```python
#unsupervized learning algorithm not a ML algorithm (just to reduce paramters)
#PCA just transform our data and attempts to find out what features explain the most variance in your data
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#working with cancer dataset

from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer() #sklearn dataset base Bunch time- acts like dictionary
```

```python
cancer.keys()
```

```python
print(cancer['DESCR']) #30 atributes, all numeric
```

```python
#want to predict whater tumor is malignant or benign
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df
```

```python
#before using any ML algoryithm it good to use PCA fist to see what features are important in predictions of tumor type
#we want to use PCA to find first two principal components and visualize it in 2D (we have 30 D with full dataset)

#first everything need to have a standard scale
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df)
scaled_data=scaler.transform(df)
```

```python
#initiatinte PCA object
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
#plot data
#all rows from colum 0 [:,0], all rows from column q [:,1]
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target']) #color by target value (malignang or bening)
plt.xlabel('Fisrt Principal Components')
plt.ylabel('Second Principal Components')
#based of just first and second PC we can see very clear separation of target feature, based of just two PC versus 30 components data
```

```python
#feature components do nto correspond 1:1 to feature data
#they correcpond to combination of selected features
pca.components_ #each row represens a principal component and to each column relates to original features
```

```python
df_comp=pd.DataFrame(pca.components_,columns=cancer['feature_names'])
df
```

```python
sns.heatmap(df_comp,cmap='plasma') 
#heatmap show the relationship btw correlation of principal components with features
#each principal components is shown as a row, the higher the number of scale the greater correlation
```
